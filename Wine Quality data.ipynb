```python
x=df.iloc[:,:-1].values
y=df.iloc[:,-1].values
print(x)
print(y)
print(x.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1,random_state=0)
print(x_train.shape)
print(x_test.shape)
```

```python
#Feature Scaling
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train_sc=sc.fit_transform(x_train)
x_test_sc=sc.fit_transform(x_test)
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
from sklearn.metrics import precision_score
```

# Implement three Models with hyperParameter tuning

```python
#Random forest

from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier(criterion='entropy',max_features=3,n_estimators=60,n_jobs=3)
rf.fit(x_train_sc,y_train)
print("Accuracy of random forest on training dataset: ",rf.score(x_train_sc,y_train))

y_pred_rf=rf.predict(x_test_sc)
rf_score=accuracy_score(y_test,y_pred_rf)
rf_results = confusion_matrix(y_test, y_pred_rf)
rf_miss = (y_test!=y_pred_rf).sum()
print('Miscalssified samples: %d'%rf_miss)
print('Accuracy score of Random forest on test datset: %.2f'%rf_score)
rf_f1=f1_score(y_test, y_pred_rf, average='micro')
rf_prec = precision_score(y_test, y_pred_rf, average="micro")
print('Precision Score: %.2f' %rf_prec)
print('F1 Score: %.2f '%rf_f1)

print("Result of Confusion matrix: \n",rf_results )
```

```python
#tuning of Hyperoarameter------------------------------------------------------------------TUNING

from sklearn.model_selection import GridSearchCV
import numpy as np

max_features_range = np.arange(1,12,1)

n_estimators_range = np.arange(10,210,10)
param_grid   = {"criterion":["gini","entropy"],"max_features":max_features_range,"n_estimators":n_estimators_range}

```

```python
rf=RandomForestClassifier()
grid = GridSearchCV(estimator=rf, param_grid=param_grid, cv=5)
```

```python
grid.fit(x_train_sc,y_train)
```

```python
print("The best parameter are %s with a score of %0.2f"
     % (grid.best_params_,grid.best_score_))
```

```python
grid_results = pd.concat([pd.DataFrame(grid.cv_results_["params"]),pd.DataFrame(grid.cv_results_["mean_test_score"], columns=["Accuracy"])],axis=1)
grid_results.head()
```

```python
#Decision tree
from sklearn.tree import DecisionTreeClassifier
tree=DecisionTreeClassifier(criterion='entropy',max_depth=3)
tree.fit(x_train_sc,y_train)
```

```python

y_pred_tree=tree.predict(x_test_sc)

print("Accuracy of Decision tree model on training set: ", tree.score(x_train_sc,y_train))
tree_results=confusion_matrix(y_test,y_pred_tree)
tree_score=accuracy_score(y_test,y_pred_tree)
tree_miss = (y_test!=y_pred_tree).sum()
print('Miscalssified samples: %d'%tree_miss)
print('Accuracy score of Decision Tree model on test set: %.2f'%tree_score)
tree_f1=f1_score(y_test, y_pred_tree, average='micro')
print('F1 Score: %.2f '%tree_f1)
tree_prec = precision_score(y_test, y_pred_tree, average="micro")
print('Precision Score: %.2f' %tree_prec)
print("Result of Confusion matrix: \n",tree_results )
```

```python

#Knn--------------------------------------------KNN-----------------------------
from sklearn.neighbors import KNeighborsClassifier
knn_scorelist = []
for k in range(21):
    k = k+1
    knn=KNeighborsClassifier(n_neighbors=k)
    knn.fit(x_train_sc,y_train)
    knn_pred=knn.predict(x_test_sc)
    knn_results = confusion_matrix(y_test,knn_pred)

    knnacc_score = accuracy_score(y_test, knn_pred)
    print("The Accuracy of Knn model for k=",k,"is:%0.4f"%knnacc_score)
    print("The confusion matrix is: \n", knn_results)
    
    knn_scorelist.append(knnacc_score)
```

```python
print("Maximum accuracy using knnmodel is: %0.4f " %max(knn_scorelist))
```
